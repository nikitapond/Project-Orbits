The overall aim of this project is the running of a simulation that calculates the positions of a set of interacting bodies as a function of time, and to then in some way display this to the user.
The basis of this project seems quite simple, and so it would be interesting to write code that can be used generally to run a simulation for a given N bodies with 
It makes sense to attempt solving for two bodies before turning to N bodies.
Let us imagine two bodies A and B, with mass' and positions $M_A,M_B, R_A,R_B$.
They exert a gravitational force on each other, and so fulfil the following equation:
$$M_A \frac{d^2R_A}{dt^2}=G\frac{\cdot M_AM_B}{|R|^3}\textbf{R}$$
Where $\textbf{R}$ is the displacement between the two objects $R_B-R_A$

With some googling i was able to find another python project based on this problem [1]
This models first a two body, then a three body system, using the alpha centauri system. I cannot find a date of publish for the article, but various functions are depricated and so had to be modified until the code ran.
After succesfully running, i re-wrote it, but with my own Runge-Kutta method to solve ODE. The code is below

Solving an ODE in python can be done by using scipy.integrate
This function requires a function that takes our second order ODE and turns it into two coupled single order ODEs
$$V_A = \frac{dR}{dt}$$
$$\frac{d^2R_A}{dt^2} = \frac{GM_B}{|R^3|}V_A$$
We can use the code from Week 4 and adapt it, we can attempt to form a simulation of the Earth, Moon, and Sun.
Below is my first initial attempt

```python
import numpy as np

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import scipy as sci
G=6.67408e-11 #m^3Kg^-1s^-1
Mass_Sun = 1.99e30 #Kg
Rel_Mass_Sun = 1# Relative to above
Rel_Mass_Earth = 3e-6 # Relative mass compared to sun in Kg
Rel_Mass_Moon = 3.7e-8 #Relative mass of moon compared to sun (in Kg)
Distance_scale = 1.5e8 #Km (1 AU)
Velocity_scale = 30000 #m/s
Time_scale = 24*3600 #s #orbital period of earth

k1 = G*Time_scale*Mass_Sun/(Distance_scale**2 * Velocity_scale)
k2 = Velocity_scale*Time_scale/Distance_scale

Sun_r0= [0,0,0] #Place sun at oragin
Earth_r0 = [1000,0,0] #Earth 1AU away
Moon_r0 = [1010, 0, 0] #These units might be too small
Sun_v0 = [0,0,0]
Earth_v0 = [0, 1, 0]
Moon_v0 = [0,0.1,0]
def CoupledODE(rv):
    """
    Solves the RHS of newtons equation for gravity.
    Inputs:
        rv - Array like, the positions and velocities of the 3 bodies, as a 1d array. Takes the form:
            [x1,y1,z1,x2,...,z3, vx1, vy1,...,vz3]
    Outputs:
        rv_tp1 - Array like, the positions and velocities of the 3 bodies at the next time step.
    """
    #Extract position of each body
    r_sun = rv[:3] #Take the first 3
    r_earth = rv[3:6]
    r_moon = rv[6:9]
    #Extract velocity of each body
    v_sun = rv[9:12]
    v_earth = rv[12:15]
    v_moon = rv[15:18]
    
    #Calcualte distance between each body
    r_earth_moon = np.linalg.norm(r_moon-r_earth)
    r_sun_earth = np.linalg.norm(r_earth-r_sun)
    r_sun_moon = np.linalg.norm(r_moon - r_sun)
    
    #Find velocity change over time step
    sun_drdt = v_sun * k2
    earth_drdt = v_earth * k2
    moon_drdt = v_moon * k2

    #Find change in position... after time step
    sun_dvdt = k1*Rel_Mass_Earth*(r_earth-r_sun)/r_sun_earth**3 + k1*Rel_Mass_Moon*(r_moon-r_sun)/r_sun_moon**3
    earth_dvdt = k1*Rel_Mass_Sun*(r_sun-r_earth)/r_sun_earth**3 + k1*Rel_Mass_Moon*(r_moon-r_earth)/r_earth_moon**3
    moon_dvdt = k1*Rel_Mass_Sun*(r_sun-r_moon)/r_sun_moon**3 + k1*Rel_Mass_Earth*(r_earth_moon-r_moon)/r_earth_moon**3
    
    #Make flat arrays ready to return for use again
    r12_der = np.concatenate((sun_drdt, earth_drdt))
    r_der = np.concatenate((r12_der, moon_drdt))
    v12_der = np.concatenate((sun_dvdt,earth_dvdt))
    v_der = np.concatenate((v12_der, moon_dvdt))

    return np.concatenate((r_der, v_der))

def RK4(function,initial, dx,N):
    """
    Performs the Runge-Kutta method on the supplied function
    Inputs:
        function - the function to perform the RK4 method to
        initial - Array like, the initial values for the problem
        dx - The seperation between each point
        N - the total number of points over which to calculate
    Outputs:
        sol - the solution after N steps
    """
    #Create empty array and set initial value
    psi = np.zeros((N, len(initial)))
    t=0
    sol[0] = initial
    #Iterate each point and Runge-Kutte
    for i in range(N-1):
        k1 = dx * function(sol[i])
        k2 = dx * function(sol[i]+0.5*k1)
        k3 = dx * function(sol[i]+0.5*k2)
        k4 = dx * function(sol[i]+k3)
        sol[i+1] = sol[i] + (k1+2*k2+2*k3+k4)/6


    return sol

#Create initial conditions array
initial =np.array([Sun_r0, Earth_r0, Moon_r0, Sun_v0, Earth_v0, Moon_v0])
initial = initial.flatten()
```

The above code now only needs to be run through the RK4 function and then plotted.

Now that I have working code to simulate the earth-moon-sun system, it makes sense to extend the code to work easily for any given 3 bodies. This is below

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy import integrate

G=6.67408e-11 #m^3Kg^-1s^-1

class Body:
    """
    class that holds details for each body
    """
    def __init__(self, name, mass, startPos, startVel):
        """
        Initiates the Body with the supplied values
        Inputs:
            name - The name of the body
            mass - The mass of the body relative to the Mass_scale supplied to
                the N body solver
            startPos - Array like (3 dimensions, [x,y,z]). Position relative to
                the Distance_scale supplied to the N body solver
            startVel - Array like (3 dimensions, [v_x, v_y, v_z]). Velocity
                relative to the Velocity_Scale supplied to the N body solver.
        """
        self.name=name
        self.mass=mass
        self.startPos=startPos
        self.startVel=startVel

def Coupled3BodyODE(rv, t, k1, k2, m1, m2, m3):
    """
    Solves the RHS of newtons equation for gravity for the 3 supplied bodies
    Inputs:
        rv - Array like, the positions and velocities of the 3 bodies, as a 1d array. Takes the form:
            [x1,y1,z1,x2,...,z3, vx1, vy1,...,vz3]
        t - time  (unused, required for compatability with scipy.odeint)
        k1 - constant calculated based on scale values, used to find velocity
            of each body
        k2 - constant calculated based on scale values, used to find acceleration
            of each body
        m1,m2,m3 - Relative masses of each body
    Outputs:
        rv_tp1 - Array like, the positions and velocities of the 3 bodies at the next time step.
    """
    #Extract position and velocity
    r1 = rv[:3] 
    r2 = rv[3:6]
    r3 = rv[6:9]

    v1 = rv[9:12]
    v2 = rv[12:15]
    v3 = rv[15:18]
    #Find distance between each body
    r_12 = np.linalg.norm(r2-r1)
    r_13 = np.linalg.norm(r3-r1)
    r_23 = np.linalg.norm(r3-r2)

    #Calculate 2 coupled differentials
    drdt_1 = v1*k2
    drdt_2 = v2*k2
    drdt_3 = v3*k2

    dvdt_1 = k1*(m2*(r2-r1)/r_12**3 + m3*(r2-r1)/r_13**3)
    dvdt_2 = k1*(m1*(r1-r2)/r_12**3 + m3*(r3-r2)/r_23**3)
    dvdt_3 = k1*(m1*(r1-r3)/r_13**3 + m2*(r2-r3)/r_23**3)
    #Flatten results and return
    r12_der = np.concatenate((drdt_1, drdt_2))
    r_der = np.concatenate((r12_der, drdt_3))
    v12_der = np.concatenate((dvdt_1,dvdt_2))
    v_der = np.concatenate((v12_der, dvdt_3))

    return np.concatenate((r_der, v_der))

def PrepairValues(bodies):
    """
    Takes a list of bodies and turns into a set of values to be used by the solver
    Inputs:
        bodies - array like, list Body objects
    Outputs:
        initial - array like, array of initial position and velocity of all bodies
        masses - array like, array of masses of all bodies.
    """
    #Prepair empty lists
    masses = []
    positions = []
    velocities =[]
    for body in bodies:
        #iterate each body and append relevent lists
        masses.append(body.mass)
        positions.append(body.startPos)
        velocities.append(body.startVel)
    #Create array of initial positions and velocities, then flatten
    initial = np.array(positions+velocities)
    initial = initial.flatten()
    #Create array of masses, then return initial values and masses
    masses = np.array(masses)
    return initial, masses

def SetRelativeValues(mass_scale, dist_scale, vel_scale, time_scale):
    """
    Calculates constants used by the differential equation solver based off
    the given values.
    Inputs:
        mass_scale (M)- The scale unit for mass (kg) that all object masses
                    are relative to
        dist_scale (D)- The scale unit for distance (m) that all distances are
                    relative to
        vel_scale (V)- The scale unit for velocity (m/s) that all velocities are
                    relative to
        time_scale (T)- The time span (s) that a value of 1 in the time span
                    used in scipy.integrare.odeint represents
    Outputs:
        k1 - The product (G*T*M)/(D^2 * V), constant to find dr/dt in ODE solver
        k2 - The product (V*T)/D, constant to find dv/dt
    """
    k1 = G*time_scale*mass_scale/(dist_scale**2 * vel_scale)
    k2 = vel_scale*time_scale/dist_scale
    return k1,k2

def SolveAndPlot3BodyProblem(bodies, k1, k2, time_span, ax=None, autoShow=True):
    """
    Prepairs the bodies supplied, then solves the 3 body problem over the given time span
    
    Inputs:
        bodies - array like, list or array length 3 of Body objects to simulate
        k1 - constant calculated based on scale values, used to find velocity
            of each body
        k2 - constant calculated based on scale values, used to find acceleration
            of each body
        time_span - Array like, time span relative to time_span set when calculating k1 and k2. 
            Time over which to simulate
        ax - Axis to plot on. Default of None. if None, create a new figure and axis to plot on.
        autoShow - boolean. If true, call plt.show() at end. If plotting onto own axisis (i.e ax!=None),
            this should be false
        
        
    """
    #Generate initial values
    initial, masses = PrepairValues(bodies)
    #Run simulation
    three_body_sol = integrate.odeint(Coupled3BodyODE, initial, time_span, args=(k1,k2,masses[0],masses[1], masses[2]))
    #Extract positions of all bodies
    b1 = three_body_sol[:,:3]
    b2 = three_body_sol[:,3:6]
    b3 = three_body_sol[:,6:9]
    #Check if axis is None, if None, generate new 3d axis
    if(ax==None):
        fig=plt.figure(figsize=(5,5))
        ax=fig.add_subplot(111,projection="3d")
    
    #Plot the orbits
    ax.plot(b1[:,0],b1[:,1],b1[:,2],color="r")
    ax.plot(b2[:,0],b2[:,1],b2[:,2],color="b")
    ax.plot(b3[:,0],b3[:,1],b3[:,2],color="g")

    #Plot the final positions of the stars
    ax.scatter(b1[-1,0],b1[-1,1],b1[-1,2],color="r",marker="o",s=100,label=bodies[0].name)
    ax.scatter(b2[-1,0],b2[-1,1],b3[-1,2],color="b",marker="o",s=100,label=bodies[1].name)
    ax.scatter(b3[-1,0],b3[-1,1],b3[-1,2],color="g",marker="o",s=100,label=bodies[2].name)
    ax.set_xlabel("x-coordinate",fontsize=14)
    ax.set_ylabel("y-coordinate",fontsize=14)
    ax.set_zlabel("z-coordinate",fontsize=14)
    ax.set_title("Visualization of a 3 body system\n",fontsize=14)
    ax.legend(loc="upper left",fontsize=14)
    
    if(autoShow):
        plt.show()
#Create list of bodies
bodies = []
bodies.append(Body("sun",1.989e6, [0,0,0], [0,0,0]))
bodies.append(Body("Earth",5.972,[145000, 0, 0], [0,30,0]))
bodies.append(Body("Moon", 7.346e-2, [145000+384, 0, 0], [0, 30+1.022, 0]))

#define scale values to use
Mass_scale = 10e23 #Kg
Distance_scale = 1000000 #m
Velocity_scale = 1000 #m/s
Time_scale = 365.25*24*3600 #s #orbital period of earth

#Calculate k1 and k2 based on scale values
k1,k2 = SetRelativeValues(Mass_scale, Distance_scale, Velocity_scale, Time_scale)


#Define time span (1 year, 500 points)
time_span=np.linspace(0,1,500)
#Solve an plot. No axis and autoShow=true, displays automatically
SolveAndPlot3BodyProblem(bodies, k1, k2, time_span)
```

The next step is to make this work for a general set of N bodies. 

```python
def CoupledNBodyODE(rv, t, k1, k2, N, masses):
    """
    Calculates the new velocity and position of each of the bodies at the given
    iteration.
    Inputs:
        rv - array like, holds position and velocity of all bodies
        t - supplied for function to work with scipy.integrate.odeint, unused
        k1 - constant calculated based on scale values, used to find velocity
            of each body
        k2 - constant calculated based on scale values, used to find acceleration
            of each body
        masses - array like, mass of each of the bodies
    Outputs - flat 1d array of floats, represents position and velocity of all
            bodies. Same format as input 'rv'
    """
    #Prepair arrays to hold positions, velocities, and position deivatives
    all_r = []
    all_v = []
    all_drdt = []
    delta_to_v = 3*N

    
    #Iterate the data set and fill arrays with required values
    for i in range(N):
        all_r.append(rv[3*i:3*(i+1)])
        v_i = rv[(3*i+delta_to_v):(3*(i+1)+delta_to_v)]
        all_v.append(v_i)
        all_drdt.append(v_i*k2)
        
    
    #Create matrix of distances between each body
    rs = np.zeros((N,N))
    for i in range(N):
        for j in range(N):
            #Any distance r_ij for j=i is 0
            if(i==j):
                continue
            #Calculate distance between bodies i and j
            rs[i,j] = np.linalg.norm(all_r[j]-all_r[i])

    #Turn masses array to flat numpy array for ease
    masses = np.array(masses).flatten()
    #Initiate velocity derivative array
    all_dvdt=[]
    #Iterate each body
    for i in range(N):
        #Initiate the velocity derivative for body i as (0,0,0), to prepair for calcualtion
        body_i_pre_mult = np.zeros(3)
        for j in range(N):
            if(j!=i):
                #Add the acceleration contribution from body j to the body i
                body_i_pre_mult += masses[j]*(all_r[j]-all_r[i])/rs[i,j]**3
        #Add total calcualted velocity change to total array
        all_dvdt.append(k1*body_i_pre_mult)

    #Turn to numpy arrays, concatenate, and flatten, then return to odeint
    all_dvdt = np.array(all_dvdt)
    return np.concatenate((all_drdt, all_dvdt)).flatten()

def SolveAndPlotNBodyProblem(bodies, k1, k2, time_span, ax=None, autoShow=True):
    initial, masses = PrepairValues(bodies)
    N = len(bodies)

    n_body_sol = integrate.odeint(CoupledNBodyODE, initial, time_span, args=(k1,k2,N,[[masses]]))

    if(ax==None):
        fig=plt.figure(figsize=(5,5))
        ax=fig.add_subplot(111,projection="3d")

    for i in range(N):
        b = n_body_sol[:,(3*i):(3*(i+1))]
        ax.plot(b[:,0],b[:,1],b[:,2], label=bodies[i].name)

    ax.set_xlabel("x-coordinate",fontsize=14)
    ax.set_ylabel("y-coordinate",fontsize=14)
    ax.set_zlabel("z-coordinate",fontsize=14)
    ax.set_title("Visualization of a system of \n" + str(N) + " gravitionally interacting bodies",fontsize=14)
    ax.legend(loc="upper left",fontsize=14)
    if(autoShow):
        plt.show()
```

```python
def CoupledNBodyODE(rv, t, k1, k2, N, masses):
    """
    Calculates the new velocity and position of each of the bodies at the given
    iteration.
    Inputs:
        rv - array like, holds position and velocity of all bodies
        t - supplied for function to work with scipy.integrate.odeint, unused
        k1 - constant calculated based on scale values, used to find velocity
            of each body
        k2 - constant calculated based on scale values, used to find acceleration
            of each body
        masses - array like, mass of each of the bodies
    Outputs - flat 1d array of floats, represents position and velocity of all
            bodies. Same format as input 'rv'
    """
    #Prepair arrays to hold positions, velocities, and position deivatives
    all_r = []
    all_drdt = []
    delta_to_v = 3*N

    
    tMass = 0
    #Iterate the data set and fill arrays with required values
    for i in range(N):
        all_r.append(rv[3*i:3*(i+1)])
        all_drdt.append(k2*rv[(3*i+delta_to_v):(3*(i+1)+delta_to_v)])
        
    
    #Create matrix of distances between each body
    rs = np.zeros((N,N))
    for i in range(N):
        for j in range(N):
            #Any distance r_ij for j=i is 0
            if(i==j):
                continue
            #rs_ij=rs_ji, prevents double calculations
            if(rs[j,i] != 0):
                rs[i,j] = rs[j,i]
            else:
                #Calculate distance between bodies i and j
                rs[i,j] = np.linalg.norm(all_r[j]-all_r[i])

    #Turn masses array to flat numpy array for ease
    masses = np.array(masses).flatten()
    #Initiate velocity derivative array
    all_dvdt=[]
    #Iterate each body
    for i in range(N):
        #Initiate the velocity derivative for body i as (0,0,0), to prepair for calcualtion
        body_i_pre_mult = np.zeros(3)
        for j in range(N):
            if(j!=i):
                #Add the acceleration contribution from body j to the body i
                body_i_pre_mult += masses[j]*(all_r[j]-all_r[i])/rs[i,j]**3
        #Add total calcualted velocity change to total array
        all_dvdt.append(k1*body_i_pre_mult)

    #Turn to numpy arrays, concatenate, and flatten, then return to odeint
    all_dvdt = np.array(all_dvdt)
    return np.concatenate((all_drdt, all_dvdt)).flatten()

def SolveAndPlotNBodyProblem(bodies, k1, k2, time_span, ax=None, autoShow=True):
    initial, masses = PrepairValues(bodies)
    N = len(bodies)

    n_body_sol = integrate.odeint(CoupledNBodyODE, initial, time_span, args=(k1,k2,N,[[masses]]))

    if(ax==None):
        fig=plt.figure(figsize=(5,5))
        ax=fig.add_subplot(111,projection="3d")

    for i in range(N):
        b = n_body_sol[:,(3*i):(3*(i+1))]
        ax.scatter(b[0,0], b[0,1], b[0,2], marker='o', color=colours[i])
        ax.plot(b[:,0],b[:,1],b[:,2], color=colours[i], label=bodies[i].name)
        ax.scatter(b[-1,0], b[-1,1], b[-1,2], marker='*', color=colours[i])


    ax.set_xlabel("x-coordinate",fontsize=14)
    ax.set_ylabel("y-coordinate",fontsize=14)
    ax.set_zlabel("z-coordinate",fontsize=14)
    ax.set_title("Visualization of a 3 body system\n",fontsize=14)
    ax.legend(loc="upper left",fontsize=14)
    if(autoShow):
        plt.show()
```

Although not required, I wanted to create code that could easily be run by others. To do this, I created a wrapper class classed NBodySolver.
This was created in a way that a user can create a new NBodySolver object, to which they simply add a set of bodies (currently limited to 7 due to the colour array having only 7 elements, but this can be easily expanded). They then run a function which solves the problem for a supplied time frame, and then another function to plot or animate the solution. I also added in easily functionality for the plot to be saved as a png file, which I used to easily and quickly create plots for my final report. The total code, as well as a commented out section at the bottom, is provided below.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
import mpl_toolkits.mplot3d.axes3d as p3
from scipy import integrate

G=6.67408e-11 #m^3Kg^-1s^-1 #Big G
"""
Below are the default values for scales of mass, distance, velocity, as well
as the time scale
"""
Mass_scale = 1e24 #Kg
Distance_scale = 1000000 #m (1e6)
Velocity_scale = 1000 #m/s (1e3)
Time_scale = 365.25*24*3600 #s #orbital period of earth

colours = ['r','g','b','y','pink', 'c', 'm'] #Colours used for plotting

def SetRelativeValues(mass_scale, dist_scale, vel_scale, time_scale):
    """
    Calculates constants used by the differential equation solver based off
    the given values.
    Inputs:
        mass_scale (M)- The scale unit for mass (kg) that all object masses
                    are relative to
        dist_scale (D)- The scale unit for distance (m) that all distances are
                    relative to
        vel_scale (V)- The scale unit for velocity (m/s) that all velocities are
                    relative to
        time_scale (T)- The time span (s) that a value of 1 in the time span
                    used in scipy.integrare.odeint represents
    Outputs:
        k1 - The product (G*T*M)/(D^2 * V), constant to find dr/dt in ODE solver
        k2 - The product (V*T)/D, constant to find dv/dt
    """
    k1 = G*time_scale*mass_scale/(dist_scale**2 * vel_scale)
    k2 = vel_scale*time_scale/dist_scale
    return k1,k2

def SetDefaultReatives():
    """
    Calculates constants used by the differential equation solver based off
    the default values.
    Default values:
        Mass_scale = 10e23 Kg, Distance_scale = 10e5 m, Velocity_scale = 1000 m/s,
        Time_scale = 365.25*24*3600 s, 1 orbit for Earth
    Outputs:
        k1 - The product (G*T*M)/(D^2 * V), constant to find dr/dt in ODE solver
        k2 - The product (V*T)/D, constant to find dv/dt
    """
    k1 = G*Time_scale*Mass_scale/(Distance_scale**2 * Velocity_scale)
    k2 = Velocity_scale*Time_scale/Distance_scale
    return k1,k2
class NBodySolver:
    def __init__(self):
        """
        Initialises the NBodySolver
        """
        #Set up list of bodies
        self.bodies = []
        #Set constants as default, set solved to False
        self.k1, self.k2 = SetDefaultReatives()
        self.solved=False

    def AddBody(self, body):
        """
        Adds the supplied body to this solver.
        Inputs:
            Body - the body to add to this solver
        """
        self.bodies.append(body)

    def AddNewBody(self, name, mass, position, velocity):
        """
        Creates a new Body based on the given arguments, and then adds it to
        this solver.
        Inputs:
            name - The name of the body
            mass - The mass of the body relative to the Mass_scale
            position - The initial position of the body relative to Distance_scale
            velocity - The initial velocity of the body relative to Velocity_scale
        """
        self.bodies.append(Body(name, mass, position, velocity))

    def SetSolverRelativeValues(self, mass_scale, dist_scale, vel_scale, time_scale):
        """
        Calculates constants used by the differential equation solver based off
        the given values.
        Inputs:
            mass_scale (M)- The scale unit for mass (kg) that all object masses
                        are relative to
            dist_scale (D)- The scale unit for distance (m) that all distances are
                        relative to
            vel_scale (V)- The scale unit for velocity (m/s) that all velocities are
                        relative to
            time_scale (T)- The time span (s) that a value of 1 in the time span
                        used in scipy.integrare.odeint represents
        """
        self.k1,self.k2 = SetRelativeValues(mass_scale, dist_scale, vel_scale, time_scale)

    def SolveNBodyProblem(self, time_span):
        """
        Prepairs the bodies of this solver, ready to be added to the ODE solver.
        Extracts the relevent data from the result, and saves to the object.
        Inputs:
            time_span - The time span that the simulation should be run over, a
                time span of 1 represents 1 Time_scale
        """
        self.time_span = time_span
        initial, masses = PrepairValues(self.bodies)
        self.N = len(self.bodies)
        n_body_sol = integrate.odeint(CoupledNBodyODE, initial, time_span, args=(self.k1,self.k2,self.N, [masses]))
        #Create array of just the positions of the solution
        self.bodySol = []
        for i in range(self.N):
            self.bodySol.append(n_body_sol[:,(3*i):(3*(i+1))])
        self.solved=True
        #Return both the neat solution, as well as the full solution
    def PlotNBodySolution(self, ax=None, show=True, saveFile=None, legend=True):
        if(self.solved==False):
            print("Solution must be found before plotting. Use NBodySolver.SolveNBodyProblem")
            return

        if(ax==None):
            fig=plt.figure(figsize=(5,5))
            ax=fig.add_subplot(111,projection="3d")
        #Iterate each body and extract full path for each body, then plot.
        for i in range(self.N):
            b = self.bodySol[i]
            ax.scatter(b[0,0], b[0,1], b[0,2], marker='o', color=colours[i])
            ax.plot(b[:,0],b[:,1],b[:,2],color=colours[i], label=self.bodies[i].name)
            ax.scatter(b[-1,0], b[-1,1], b[-1,2], marker='*', color=colours[i])
        #Add details to plot, then show
        ax.set_xlabel("x-coordinate",fontsize=10)
        ax.set_ylabel("y-coordinate",fontsize=10)
        ax.set_zlabel("z-coordinate",fontsize=10)
        ax.set_title("Visualization of a 3 body system\n",fontsize=14)
        if(legend):
            ax.legend(loc="lower left",fontsize=10)
        if(show):
            plt.show()
        if(saveFile != None):
            fig.savefig(saveFile, dpi=fig.dpi)
    def AnimateNBodySolution(self, axis_size=None):
        if(self.solved==False):
            print("Solution must be found before animating. Use NBodySolver.SolveNBodyProblem")
            return
        data = []
        for i in range(self.N):
            data.append([self.bodySol[i][:,0],self.bodySol[i][:,1],self.bodySol[i][:,2]])
        #Turn to numpy array
        data = np.array(data)
        #Check if axis_size is defined. If not, define it
        if(axis_size==None):
            axis_size = np.max(self.bodySol)
        #Create 3d figure to plot on
        fig=plt.figure(figsize=(6,6))
        ax=fig.add_subplot(111,projection="3d")
        #Extract data into a set of 3 dimensional lines
        lines = [ax.plot(dat[0,0:1],dat[1,0:1],dat[2,0:1], label=self.bodies[i].name)[0] for i, dat in enumerate(data)]
        for i, line in enumerate(lines):
            line.set_color(colours[i])

            #line.label("test" + str(i))

        def update_lines(num, dataLines, lines):
            """
            Update function for the animation.
            Inputs:
                num - the current iteration of the animation
                dataLines - all of the 3d position solutions of all bodies
                lines - the lines to animate
            """

            #i=0
            for line, data in zip(lines, dataLines):
                #line.set_color(colours[i])
                line.set_data(data[0:2, :num])
                line.set_3d_properties(data[2, :num])

                #i+=1
            return lines
        ax.legend(loc="lower left",fontsize=14)
        #Set up axis of plot
        ax.set_xlim3d([-axis_size,axis_size])
        ax.set_xlabel('X')

        ax.set_ylim3d([-axis_size,axis_size])
        ax.set_ylabel('Y')

        ax.set_zlim3d([-axis_size,axis_size])
        ax.set_zlabel('Z')
        #Create animation, then show to user
        line_ani = FuncAnimation(fig, update_lines, len(self.time_span), fargs=(data, lines),
                                       interval=0.1, blit=True, repeat=False)
        plt.show()

class Body:
    """
    class that holds details for each body
    """
    def __init__(self, name, mass, startPos, startVel):
        """
        Initiates the Body with the supplied values
        Inputs:
            name - The name of the body
            mass - The mass of the body relative to the Mass_scale supplied to
                the N body solver
            startPos - Array like (3 dimensions, [x,y,z]). Position relative to
                the Distance_scale supplied to the N body solver
            startVel - Array like (3 dimensions, [v_x, v_y, v_z]). Velocity
                relative to the Velocity_Scale supplied to the N body solver.
        """
        self.name=name
        self.mass=mass
        self.startPos=startPos
        self.startVel=startVel


def CoupledNBodyODE(rv, t, k1, k2, N, masses):
    """
    Calculates the new velocity and position of each of the bodies at the given
    iteration.
    Inputs:
        rv - array like, holds position and velocity of all bodies
        t - supplied for function to work with scipy.integrate.odeint, unused
        k1 - constant calculated based on scale values, used to find velocity
            of each body
        k2 - constant calculated based on scale values, used to find acceleration
            of each body
        masses - array like, mass of each of the bodies
    Outputs - flat 1d array of floats, represents position and velocity of all
            bodies. Same format as input 'rv'
    """
    #Prepair arrays to hold positions, velocities, and position deivatives
    all_r = []
    all_drdt = []
    delta_to_v = 3*N


    #Iterate the data set and fill arrays with required values
    for i in range(N):
        all_r.append(rv[3*i:3*(i+1)])
        all_drdt.append(rv[(3*i+delta_to_v):(3*(i+1)+delta_to_v)]*k2)

    #Convert to numpy arrays for efficiency and ease
    all_r = np.array(all_r)
    all_drdt = np.array(all_drdt)


    

    
    #Create matrix of distances between each body
    rs = np.zeros((N,N))
    for i in range(N):
        for j in range(N):
            #Any distance r_ij for j=i is 0
            if(i==j):
                continue
            #rs_ij=rs_ji, prevents double calculations
            if(rs[j,i] != 0):
                rs[i,j] = rs[j,i]
            else:
                #Calculate distance between bodies i and j
                rs[i,j] = np.linalg.norm(all_r[j]-all_r[i])

    #Turn masses array to flat numpy array for ease
    masses = np.array(masses).flatten()
    
    #Initiate velocity derivative array
    all_dvdt=[]
    #Iterate each body
    for i in range(N):
        #Initiate the velocity derivative for body i as (0,0,0), to prepair for calcualtion
        body_i_pre_mult = np.zeros(3)
        for j in range(N):
            if(j!=i):
                #Add the acceleration contribution from body j to the body i
                body_i_pre_mult += masses[j]*(all_r[j]-all_r[i])/rs[i,j]**3
        #Add total calcualted velocity change to total array
        all_dvdt.append(k1*body_i_pre_mult)

    #Turn to numpy arrays, concatenate, and flatten, then return to odeint
    all_dvdt = np.array(all_dvdt)
    return np.concatenate((all_drdt, all_dvdt)).flatten()

def PrepairValues(bodies):
    """
    Takes a list of bodies and turns into a set of values to be used by the solver
    Inputs:
        bodies - array like, list Body objects
    Outputs:
        initial - array like, array of initial position and velocity of all bodies
        masses - array like, array of masses of all bodies.
    """
    #Prepair empty lists
    masses = []
    positions = []
    velocities =[]
    #iterate each body and append relevent lists
    for body in bodies:
        masses.append(body.mass)
        positions.append(body.startPos)
        velocities.append(body.startVel)
    #Create array of initial positions and velocities, then flatten
    initial = np.array(positions+velocities)
    initial = initial.flatten()
    #Create array of masses, then return initial values and masses
    masses = np.array(masses)
    return initial, masses




# Code Use
# NBodyPlotter.py is designed such that the user can easily simulated any n body
# system, and either plot or animate.
# Example code commented out shows use of code



# #Initialises the solver with default values
# solver = NBodySolver()

# #Add 4 bodies to the solver with iniial starting conditions
# solver.AddBody(Body("sun",1.989e6, [0,0,0], [0,0,0]))
# solver.AddBody(Body("earth",5.972,[148070, 0, 0], [0,29.72,0]))
# solver.AddBody(Body("mecruary", 3.285e-1, [52868,0,0], [0,48,0]))
# solver.AddBody(Body("venus", 4.867, [107630, 0, 0], [0,35,0]))
# #Define a time span of 5 solar years, with 12000 data points total
# time_span=np.linspace(0,5,12000)

# #Solver the problem over the time span, and save to "test.png"
# solver.SolveNBodyProblem(time_span)
# solver.PlotNBodySolution(saveFile="test.png")
```

# 26/2/20
Upon meeting my mini-project, we discussed my code so far. She suggested I add in some code that results in the plots being in the centre of mass reference frame. This can be done easily by first finding the centre of mass:
$$R_{CM}=\frac{\sum_i r_i\times m_i }{\sum_i m_i}$$
And then subtracting this value from the position of each body at every iteration. The modified function is below.

```python


def CoupledNBodyODE(rv, t, k1, k2, N, masses):
    """
    Calculates the new velocity and position of each of the bodies at the given
    iteration relative to the centre of mass of the system
    Inputs:
        rv - array like, holds position and velocity of all bodies
        t - supplied for function to work with scipy.integrate.odeint, unused
        k1 - constant calculated based on scale values, used to find velocity
            of each body
        k2 - constant calculated based on scale values, used to find acceleration
            of each body
        masses - array like, mass of each of the bodies
    Outputs - flat 1d array of floats, represents position and velocity of all
            bodies. Same format as input 'rv'
    """
    #Prepair arrays to hold positions, velocities, and position deivatives
    all_r = []
    all_drdt = []
    delta_to_v = 3*N
    
     #Turn masses array to flat numpy array for ease
    masses = np.array(masses).flatten()
    tMass = np.sum(masses)
    #Initiate centre of mass
    cm = np.array([0,0,0])


    #Iterate the data set and fill arrays with required values
    for i in range(N):
        all_r.append(rv[3*i:3*(i+1)])
        all_drdt.append(rv[(3*i+delta_to_v):(3*(i+1)+delta_to_v)]*k2)
        cm=np.add(cm,all_drdt[i]*masses[i])
    #Convert to numpy arrays for efficiency and ease
    all_r = np.array(all_r)
    all_drdt = np.array(all_drdt)
    cm/=tMass
    #Subtract CM from each position.
    for i in range(N):
        all_drdt[i] -= cm
    

    
    #Create matrix of distances between each body
    rs = np.zeros((N,N))
    for i in range(N):
        for j in range(N):
            #Any distance r_ij for j=i is 0
            if(i==j):
                continue
            #rs_ij=rs_ji, prevents double calculations
            if(rs[j,i] != 0):
                rs[i,j] = rs[j,i]
            else:
                #Calculate distance between bodies i and j
                rs[i,j] = np.linalg.norm(all_r[j]-all_r[i])

   
    
    #Initiate velocity derivative array
    all_dvdt=[]
    #Iterate each body
    for i in range(N):
        #Initiate the velocity derivative for body i as (0,0,0), to prepair for calcualtion
        body_i_pre_mult = np.zeros(3)
        for j in range(N):
            if(j!=i):
                #Add the acceleration contribution from body j to the body i
                body_i_pre_mult += masses[j]*(all_r[j]-all_r[i])/rs[i,j]**3
        #Add total calcualted velocity change to total array
        all_dvdt.append(k1*body_i_pre_mult)

    #Turn to numpy arrays, concatenate, and flatten, then return to odeint
    all_dvdt = np.array(all_dvdt)
    return np.concatenate((all_drdt, all_dvdt)).flatten()
```

```python
#Define scale values to keep close to unity
mass_scale = 1.5e29 #Kg
dist_scale = 1e11 #m
vel_scal = 1000 #m/s (such that inputted units are in Km/s)
orbit_period = 356*24*60*60 #s

t = 20
time_span=np.linspace(0,t,t*1000)
#Create figures
fig = plt.figure(figsize=(20,8))
ax1 = fig.add_subplot(151, projection='3d')
ax2 = fig.add_subplot(152, projection='3d')
ax3 = fig.add_subplot(153, projection='3d')
ax4 = fig.add_subplot(154, projection='3d')
ax5 = fig.add_subplot(155, projection='3d')

#Initiate solver
solver = NBodySolver()
solver.SetSolverRelativeValues(mass_scale, dist_scale, vel_scal, orbit_period)

#Define initial values for 3 bodies
ip1 = [0.97000436, -0.24308753, 0]
ip2 = [-ip1[0], -ip1[1], 0]
ip3 = [0,0,0]
iv3 = [-0.93240737*10,-0.86473146*10, 0]
iv2 = [-iv3[0]/2, -iv3[1]/2, 0]
iv1 = np.copy(iv2)

#Create and plot stable figure of 8 orbit
solver.AddBody(Body("body1",1, ip1, iv1))
solver.AddBody(Body("body2",1, ip2, iv2))
solver.AddBody(Body("body3",1, ip3, iv3))

solver.SolveNBodyProblem(time_span)
solver.PlotNBodySolution(ax=ax1, show=False)
ax1.set_title("Stable figure of 8 orbit\n")

solver.bodies.clear()
#Create and plot figure of 8 with slight change to v3
solver.AddBody(Body("body1",1, ip1, iv1))
solver.AddBody(Body("body2",1, ip2, iv2))
solver.AddBody(Body("body3",1, ip3, np.array(iv3)*1.1))

solver.SolveNBodyProblem(time_span)
solver.PlotNBodySolution(ax=ax2, show=False, legend=False)
ax2.set_title(r"$v_{3_{new}} =1.1v_{3}$ - Stable")

solver.bodies.clear()
#Create and plot figure of 8 with larger change to v3
solver.AddBody(Body("body1",1, ip1, iv1))
solver.AddBody(Body("body2",1, ip2, iv2))
solver.AddBody(Body("body3",1, ip3, np.array(iv3)*1.2))

solver.SolveNBodyProblem(time_span)
solver.PlotNBodySolution(ax=ax3, show=False, legend=False)
ax3.set_title(r"$v_{3_{new}} =1.2v_{3}$ - Chaotic")

solver.bodies.clear()
#Create and plot figure of 8 with slight change to m3
solver.AddBody(Body("body1",1, ip1, iv1))
solver.AddBody(Body("body2",1, ip2, iv2))
solver.AddBody(Body("body3",1.1, ip3, iv3))

solver.SolveNBodyProblem(time_span)
solver.PlotNBodySolution(ax=ax4, show=False, legend=False)
ax4.set_title(r"$m_{3_{new}} =1.1m_{3}$ - Stable")

solver.bodies.clear()
#Create and plot figure of 8 with larger change to m3
solver.AddBody(Body("body1",1, ip1, iv1))
solver.AddBody(Body("body2",1, ip2, iv2))
solver.AddBody(Body("body3",1.2, ip3, iv3))

solver.SolveNBodyProblem(time_span)
solver.PlotNBodySolution(ax=ax5, show=False, legend=False)
ax5.set_title(r"$m_{3_{new}} =1.2m_{3}$ - Chaotic")
fig.suptitle("Varying initial masses and velocities of a stable 3-body system")
```

We see that it only takes a 20% change in either the mass or velocity of a given body to go from a completely stable system to a completely chatoic one.

### Butterfly
This 3 body system was found by a group of scientists who used a super computer to find stable orbits, based again on the idea of symmetry [3]. I chose to take 1 specific orbit, the butterfly orbit. I then, as before, ran two seperate experiments to check the stability of the system with different starting conditions, and with the addition of an extra body.


```python

#Define scale values to keep close to unity
mass_scale = 1e30 #Kg
dist_scale = 1e11 #m
vel_scal = 1000 #m/s (such that inputted units are in Km/s)
orbit_period = 356*24*60*60 #s

solver = NBodySolver()
solver.SetSolverRelativeValues(mass_scale, dist_scale, vel_scal, orbit_period)


#Create figures
fig = plt.figure(figsize=(20,8))
ax1 = fig.add_subplot(151, projection='3d')
ax2 = fig.add_subplot(152, projection='3d')
ax3 = fig.add_subplot(153, projection='3d')
ax4 = fig.add_subplot(154, projection='3d')
ax5 = fig.add_subplot(155, projection='3d')

t = 10
time_span=np.linspace(0,t,t*10000)


#Define starting positions and velocities required for orbit
ip1 = np.array([-1,0,0])
ip2 = -ip1
ip3 = np.array([0,0,0])

vx1 = 0.30689 * 10
vy1 = 0.12551 * 10
iv1 = np.array([vx1,vy1, 0])
iv2 = np.array([vx1, vy1, 0])
iv3 = np.array([-2*vx1, -2*vy1, 0])

#Plot stable system
solver.AddBody(Body("1", 0.1, ip1, iv1))
solver.AddBody(Body("2", 0.1, ip2, iv2))
solver.AddBody(Body("3", 0.1, ip3, iv3))

solver.SolveNBodyProblem(time_span)
solver.PlotNBodySolution(ax=ax1, show=False)
ax1.set_title("Stable butterfly system (" + str(t) + " years)")

#Create and plot system with slight change in initial velocity of 1 body 
solver.bodies.clear()
solver.AddBody(Body("1", 0.1, ip1, iv1*1.5))
solver.AddBody(Body("2", 0.1, ip2, iv2))
solver.AddBody(Body("3", 0.1, ip3, iv3))
solver.SolveNBodyProblem(time_span)
solver.PlotNBodySolution(ax=ax2, show=False, legend=False)
ax2.set_title(r"$v_{1_{new}}=1.5v_1$ - Stable (" + str(t) + " years)")

#Create and plot system with a larger change in initial velocity of 1 body 
solver.bodies.clear()
solver.AddBody(Body("1", 0.1, ip1, iv1*3))
solver.AddBody(Body("2", 0.1, ip2, iv2))
solver.AddBody(Body("3", 0.1, ip3, iv3))
solver.SolveNBodyProblem(time_span)
solver.PlotNBodySolution(ax=ax3, show=False, legend=False)
ax3.set_title(r"$v_{1_{new}}=3v_1$ - Chaotic (" + str(t) + " years)")

#Run next simulations over longer period, as it was unclear if simulation 3 would remain stable over time
t = 30
time_span=np.linspace(0,t,t*10000)

#Create and plot system with a slight change in initial mass of 1 body 
solver.bodies.clear()
solver.AddBody(Body("1", 0.15, ip1, iv1))
solver.AddBody(Body("2", 0.1, ip2, iv2))
solver.AddBody(Body("3", 0.1, ip3, iv3))
solver.SolveNBodyProblem(time_span)
solver.PlotNBodySolution(ax=ax4, show=False, legend=False)
ax4.set_title(r"$m_{1_{new}}=1.5m_1$ - Stable (" + str(t) + " years)")
solver.bodies.clear()

#Create and plot system with a larger change in initial mass of 1 body 
t = 10
time_span=np.linspace(0,t,t*10000)
solver.AddBody(Body("1", 0.18, ip1, iv1))
solver.AddBody(Body("2", 0.1, ip2, iv2))
solver.AddBody(Body("3", 0.1, ip3, iv3))
solver.SolveNBodyProblem(time_span)
solver.PlotNBodySolution(ax=ax5, show=False, legend=False)
ax5.set_title(r"$m_{1_{new}}=1.8m_1$ - Chaotic (" + str(t) + " years)")

fig.suptitle("Stability of Butterfly system with varying initial velocities and masses")
```

These results are quite interesting. It seems as though a change to the initial velocty (so long as its a straight scaling of the initial velocity) requires a signficant increase (300% increase) before instability is seen. 
However, increasing the mass of 1 body by only 80% causes instability. This is likely due to the fact that an increase in velocity causes a lopsided system, but the overall interactions are still roughly equal. Whereas with an increase in mass, we produce an assymettry which increases over time as each body passes close to each other and interacts.

### Halo system
This is a system I created after playing around for a bit. It consists of 2 roughly solar mass stars orbiting each other about a common circular orbit.
In the centre we place another star, with an initial velocity perpendicular to the orbital plane of the other two stars. This system attempts to remain stable through extreme symmetry, and as a result it was important to get the circular orbit of the initial two stars perfect. To do this, a relativly large orbit was chosen of $1\times 10^{11}m$, with solar like masses of $1\times 10^{30}kg$ for the two central stars. After some playing around we managed to produce an orbit stable for at least 100 years

```python
#Define scale values to keep close to unity
mass_scale = 1e30 #Kg
dist_scale = 1e11 #m
vel_scal = 1000 #m/s (such that inputted units are in Km/s)
orbit_period = 356*24*60*60 #s



star_vel = np.sqrt(G * mass_scale/(dist_scale))/(vel_scal*2)

t = 30
time_span=np.linspace(0,t,t*1000)
#Initiate solver
solver = NBodySolver()
solver.SetSolverRelativeValues(mass_scale, dist_scale, vel_scal, orbit_period)
solver.AddBody(Body("star 1", 1, [-1, 0, 0], [0,star_vel,0]))
solver.AddBody(Body("star 2", 1, [1, 0, 0], [0,-star_vel,0]))
solver.SolveNBodyProblem(time_span)
solver.PlotNBodySolution()
```

I had collected a lot of data, and made a lot of plots, and then started to write my formal report for this project.
At this point, I realised that my experiments hadn't been very scientific, and so I decided I needed to produce a quantative measurement of some sort.
I decided this measurement should be linked to what I'd already been doing, both to save time and effort.
I came up with a a value I called "rogue star approach value", labeled as $R_{rs}$.
What does this number represent?
We take a body whos mass is equal to the total mass of the system we are testing, and we set it on a trajectory such that at its closest approach it is a distance R from the systems centre, with its velocity at this point being 2 times the escape velocity (calculated by treating the system as a stationary point object). We then divide this distance by the distance of the body with the furthest initial position from the systems centre, to get the 'rogue star approach value'
This drastically reduced the number of figures required in my final paper, as now we shall only require 3 per system. The first representing the system without the rogue star (in its stable state), the second showing a value of $R_{rs}$ just before the value at which the system devolves into a chaotic state, and a third showing the system with the value of $R_{rs}$ such that the system becomes chaotic.
